```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread('snow.png')

# Convert the image to HSV color space
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Display the V channel and its histogram
v_channel = hsv_img[:, :, 2]
plt.subplot(121)
plt.imshow(v_channel, cmap='gray')
plt.title('V Channel')
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.hist(v_channel.ravel(), 256, [0, 256])
plt.title('V Channel Histogram')
plt.show()

# Apply histogram equalization to the V channel
v_channel_eq = cv2.equalizeHist(v_channel)

# Display the histogram of the equalized V channel
plt.hist(v_channel_eq.ravel(), 256, [0, 256])
plt.title('Equalized V Channel Histogram')
plt.show()

# Create a new HSV image using the original H and S channels and the equalized V channel
hsv_img_eq = cv2.merge((hsv_img[:, :, 0], hsv_img[:, :, 1], v_channel_eq))

# Convert the new HSV image back to RGB color space
img_eq = cv2.cvtColor(hsv_img_eq, cv2.COLOR_HSV2BGR)

# Display the final result
plt.imshow(cv2.cvtColor(img_eq, cv2.COLOR_BGR2RGB))
plt.title('Result')
plt.xticks([])
plt.yticks([])
plt.show()
```
